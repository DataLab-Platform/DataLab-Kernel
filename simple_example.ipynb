# DataLab-Kernel Simple Example

This notebook demonstrates the basic usage of **DataLab-Kernel**, a Jupyter kernel built on [Xeus-Python](https://github.com/jupyter-xeus/xeus-python) that provides seamless access to DataLab workspaces.

DataLab-Kernel works in both **native Jupyter** (JupyterLab, VS Code) and **[JupyterLite](https://jupyterlite.readthedocs.io/)** (browser-based, no server required).

### 1. Load the Extension

First, load the DataLab-Kernel extension to access the `workspace`, `plotter`, and `sigima` APIs:

```python
%load_ext datalab_kernel
```

### 2. Explore the Workspace

List all objects currently available in the workspace:

```python
workspace.list()
```

### 3. Retrieve and Visualize Data

Get an image from the workspace and display it inline:

```python
img = workspace.get("i001")
plotter.plot(img)
```

### 4. Process with Sigima

Use [Sigima](https://sigima.readthedocs.io/) to compute statistics and perform analysis on the image:

```python
import sigima.proc.image as spi
spi.stats(img)
```

```python
spi.enclosing_circle(img)
```
